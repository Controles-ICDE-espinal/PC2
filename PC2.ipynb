```python
import pandas as pd
import numpy as np
paginadata = "https://raw.githubusercontent.com/Controles-ICDE-espinal/PC2/refs/heads/main/DATA/HDI-TABLA.csv"
data = pd.read_csv(paginadata)
```

```python
data
```

```python
data.info()
```

```python
#se puede obeservar que en ciertas columnas hay menos valores numericos, entonces limpiamos#
data=data.drop(["Unnamed: 3","Unnamed: 5", "Unnamed: 7","Unnamed: 9","Unnamed: 11", "Unnamed: 13"], axis=1)
data
```

```python
#Se editan los subtitulos segun lo requerido primero HDI y Country y luego las demÃ¡s#
sub_nombre=list(data.iloc[4])[0:2]+list(data.iloc[3])[2:]
data.sub= sub_nombre
data.columns=sub_nombre
data
```

```python
# Se procede a cortar las lista hasta Somalia, el Ãºltimo paÃ­s de la lista.#
data = data.iloc[6:206]
data
```

```python
#Borramos las filas que contengan en la columna 2 un texto con solo MAYÃšSCULAS"
data.iloc[:, 1] = data.iloc[:, 1].astype(str)
data = data[~data.iloc[:, 1].str.isupper()]
data
```

```python
#se corrige los subtitulos del datframe#
columnas = sub_nombre

for i in range(len(columnas)):
    if i < 8: 
        columnas[i] += " (2021)"
    elif i == len(columnas) - 1:  
        columnas[i] += " (2020)"

data.columns = columnas
data
```

```python
# La columna 7(GROSS NATIONAL) se observa valores con ".", se procede a cambiar por ","#
data.iloc[:, 6] = data.iloc[:, 6].astype(str).str.replace('.', ',', regex=False)
#Por Ãºltimo los valores con ".." los pasamos a str y los remplazamos por vacÃ­o#
for col in data.columns[1:]:  
    data.loc[:, col] = data[col].astype(str).replace({"..": "", ",,": ""})
data
```

```python
data.to_csv('HDI-tabla1-valores_corregidos.csv', index=True)
```
